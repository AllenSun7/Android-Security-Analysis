# Combine tables

## Category -> Binary

```python
for col in df_plugin_usage.columns[1:]:
    df_plugin_usage[col].values[df_plugin_usage[col] > 1] = 1
```

## Change name

```python
ori_cols = df_plugin_usage.columns[1:]
ori_cols
```

```python
def add_prefix(df, prefix):
    new_cols = {c: f"{prefix}_{c}" for c in ori_cols}
#     print(new_cols)
    df_rename = df.rename(columns=new_cols)
    return df_rename
```

```python
# add_prefix to plugin_usage
df_plugin_usage_rename = add_prefix(df_plugin_usage, "u")
df_plugin_usage_rename
```

```python
# add_prefix to plugin_declaration
df_plugin_declaration_rename = add_prefix(df_plugin_declaration, "d")
df_plugin_declaration_rename
```

```python
# add_prefix to plugin_permission
df_plugin_permission_declaration_rename = add_prefix(df_plugin_permission_declaration, "p")
df_plugin_permission_declaration_rename
```

## Combine tables

```python
from functools import reduce
dfs = [df_plugin_usage_rename, df_plugin_declaration_rename, df_plugin_permission_declaration_rename, df_permission]
df_plugin = reduce(lambda x, y: pd.merge(x, y, on="apk_name", how="outer"), dfs)
df_plugin
```

```python
df_plugin.info()
```

```python
# Convert string to int
from collections import Counter
plugins_cols = df_plugin.columns[1:]
for col in plugins_cols:
    df_plugin[col] = df_plugin[col].astype(int)
    # print exception that contains more than 2 values
    values = Counter(df_plugin[col])
    if len(values) > 2:
        print(values, col)
df_plugin
```
